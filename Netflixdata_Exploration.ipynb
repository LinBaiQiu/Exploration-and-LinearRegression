## Convert the duration column values to integer format, and override the values of the original columns.

Example -> 90 min(string) -> 90(integer)

```python
dataset['duration']=dataset['duration'].fillna(0)  #replaces NAN values with 0
dataset['duration']
```

```python
dataset['duration'] = dataset['duration'].replace('[^0-9\.]','', regex=True)
dataset['duration'].fillna(0)
```

```python
dataset['duration'] = dataset['duration'].astype(int)
dataset['duration']
```
